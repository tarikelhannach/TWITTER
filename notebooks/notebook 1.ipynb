```python
def process_string(texto):
    text = re.sub(r"\@\w+[,]|\@\w+|[,]\@\w+", "", texto)
    return texto
```

```python
tweets['Texto'] = tweets['Texto'].apply(lambda x: process_string(x))
```

```python
def clean_emoji(text):
    ''' 
    Funcion para limpiar los emojis que aparecen dentro de un texto.
    
    Args:
        text (str): texto sobre el que se pretende realizar la función.
    
    Returns:
        emoj_text (text): el texto que se introduce en el argumento pero 
        sin ningún emoji.
    '''
    emoji_text = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)

    return emoji_text.sub(r'', text)
```

```python
tweets['Text'] = tweets['Texto'].apply(lambda x:clean_emoji(x))
```
